```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sensor_data_600.txt", delimiter=" ", header = None, names = ("date","time","ir","lidar"))
```

```python
plt.grid()
plt.hist(df.lidar, bins = max(df.lidar) - min(df.lidar))
```

```python
df.lidar.plot()
```

```python
df['hour'] = [e//10000 for e in df.time]
d = df.groupby('hour')
d.lidar.mean().plot()
```
